```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('sales_eda_data.csv')
```

```python
data.info()
print(data.head())
print(data.isna().sum().sort_values(ascending=False))
print(data.duplicated().sum())
data.dtypes
```

```python
data['Sale_Date'] = pd.to_datetime(data['Sale_Date'])
data['Month'] = data['Sale_Date'].dt.month
data['Year'] = data['Sale_Date'].dt.year
data['Quarter'] = data['Sale_Date'].dt.quarter
data['Product_ID'] = data['Product_ID'].astype('object')
data = data.rename(columns={'Region': 'State'})
```

```python
product_performance = data.groupby('Product_ID')['Sales_Amount'].sum().sort_values(ascending = False).reset_index()
state_product_performance = data.groupby(['State', 'Product_ID'])['Sales_Amount'].sum().sort_values(ascending = False). reset_index()

top_10_product = product_performance.head(10)
bottom_10_product = product_performance.tail(10)
top_10_state_products = state_product_performance.head(20)
bottom_10_state_products = state_product_performance.tail(10)
```

## Which products generate the highest total revenue across the business?

```python
plt.barh(top_10_product['Product_ID'].astype('str'), top_10_product['Sales_Amount'])
plt.xlabel('Total Sales Amount')
plt.ylabel('Product ID')
plt.title('Top 10 Products by Sales')
plt.gca().invert_yaxis()
plt.show()
plt.clf()

#This graph shows us the top 10 best-performing products across all our stores. We can see product 1099 with the best performance, significantly ahead of the other top-performing products.

```

## Top performing products by state?


```python
for state in state_product_performance['State'].unique():
    subset = state_product_performance[state_product_performance['State'] == state]
    top_5 = subset.head(5)
    plt.barh(top_5['Product_ID'].astype(str),top_5['Sales_Amount'])
    plt.title(f"Top Products in {state}")
    plt.xlabel("Total Sales Amount")
    plt.ylabel("Product ID")
    plt.gca().invert_yaxis()
    plt.show()

# This graph shows us our top-performing products by state. We can see a diverse product catalogue of top-performing products for each state. This is a positive as we have a range of products doing well and can analyse state promotional techniques to apply them in other states then.

```

```python
state_performance = data.groupby('State')['Sales_Amount'].sum().sort_values(ascending = False).reset_index()
print(state_performance)
plt.pie(state_performance['Sales_Amount'], labels = state_performance['State'], autopct='%1.1f%%')
plt.axis('equal')
plt.show()
plt.clf()
# From our pie chart, we can see that performance across the states is fairly even.
```

## How is revenue distributed across product categories within each state?


```python
product_category_by_state = data.groupby(['State', 'Product_Category'])['Sales_Amount'].sum().sort_values(ascending = False).reset_index()
print(product_category_by_state)

for state in product_category_by_state['State'].unique():
    subset = product_category_by_state[product_category_by_state['State'] == state]
    plt.pie(subset['Sales_Amount'], labels = subset['Product_Category'], autopct='%1.1f%%')
    plt.title(f'Sales revenue proportion per product category for {state}')
    plt.show()

# We can see a fairly even spread of the proportion of revenue generated through each product category across the 4 states. Food in QLD should be investigated as to why it is lower compared to other states and other categories within QLD. Furniture in NSW should also be looked at, as it is slightly lower compared to other states.           


Food_qld = data[data['State'] == 'QLD']
Food_qld = Food_qld.groupby('Product_Category')['Sales_Amount'].sum().reset_index()

print(Food_qld)

plt.pie(Food_qld['Sales_Amount'], labels = Food_qld['Product_Category'], autopct='%1.1f%%')
plt.show()
plt.clf()

# The lower proportion of transactions at 18.9% suggest that lower Food revenue is driven by lower volume sold. We have no supporting data to further investigate this. 
# The next steps would be to identify if the decrease in Food transactions is isolated to a specific area or if it is a general trend across the state. If isolated to a specific area, the team needs to identify what promotional campaigns are being run in these areas and how they differ from those performing well. We also need to analyse inventory levels in this area to understand if the levels are adequate for the expected demand. 
# If this is a statewide trend, we can analyse metrics such as the cost of food products in QLD compared to other states, we can analyse which products in the food category are performing well in other states and if they are being supplied in QLD and to what level. Understanding which products within the food category in QLD are performing well and not so well is important, so we can keep a product catalogue that supports the business revenue.  We also need to understand if there have been any environmental events within the state that are impacting consumers' purchasing behaviours.    

```

## How do sales revenue and transaction volumes change over time?
## Is revenue growth driven more by transaction volume or by higher-value transactions?


```python
# 2023 Sales analysis
sales_2023 = data[data['Year'] == 2023]

monthly_sales = sales_2023.groupby('Month')['Sales_Amount'].sum().reset_index().sort_values('Month')
plt.plot(monthly_sales['Month'], monthly_sales['Sales_Amount'])
plt.title('Monthly Revenue')
plt.xticks(range(1,13))
plt.show()
plt.clf()
print(monthly_sales)

monthly_transactions = sales_2023.groupby('Month')['Sales_Amount'].count().reset_index().sort_values('Month')
plt.plot(monthly_transactions['Month'], monthly_transactions['Sales_Amount'])
plt.title('Monthly Transactions')
plt.xticks(range(1,13))
plt.show()
plt.clf()
print(monthly_sales)

monthly_avg_transaction = (sales_2023.groupby('Month')['Sales_Amount'].mean().reset_index())
sns.lineplot(x = monthly_avg_transaction['Month'], y = monthly_avg_transaction['Sales_Amount'])
plt.title('Average Revenue Amount per Transaction')
plt.show()
plt.clf()


quarterly_sales = sales_2023.groupby('Quarter')['Sales_Amount'].sum().reset_index().sort_values('Quarter')
plt.plot(quarterly_sales['Quarter'], quarterly_sales['Sales_Amount'])
plt.title('Quarterly Revenue')
plt.xticks([1, 2, 3, 4])
plt.show()
plt.clf()
print(quarterly_sales)

quarterly_transactions = sales_2023.groupby('Quarter')['Sales_Amount'].count().reset_index().sort_values('Quarter')
plt.plot(quarterly_transactions['Quarter'], quarterly_transactions['Sales_Amount'])
plt.title('Quarterly Transactions')
plt.xticks([1, 2, 3, 4])
plt.show()
plt.clf()
print(quarterly_sales)

# Monthly and quarterly sales are highly variable. There is no steady trend in sales, either monthly or quarterly. January, October, and November were our best-performing months revenue-wise. Monthly sales projections could produce unreliable results with such variability. Factors such as seasonality, promotional activities or external factors should be investigated. The performance of January, October and November is a result of volume as our transactions per month graph shows spikes in these months, and we can confirm this by the low average sale per transaction shown in the following graph.
# The variability in transaction count across all months in 2023 can suggest the variability in revenue for each month. April and July should be looked at as the proportion of revenue per transaction is high, we should look into which promotions the business ran and which products and states performed well, as it clearly spiked our revenue. However, we also need to consider why transactions were so low and if that is a result of lower foot traffic.
# We see a large increase in sales from quarter 3 to 4. The quarterly transaction count supports the increase we see in revenue from quarter 3 to 4. Total transactions in quarter 4 of 2023 were at an all-time high. It suggests that the increase in revenue was more volume-based than price-driven. This could be influenced by factors previously stated, as well as product relevance with the seasonal climate.

product_performance_april = sales_2023[sales_2023['Month'] == 4]
product_performance_april = product_performance_april.groupby('Product_ID')['Sales_Amount'].sum().sort_values(ascending = False).reset_index()
top_products = product_performance_april.head(10)
sns.barplot(x = top_products['Product_ID'], y = top_products['Sales_Amount'])
plt.title('Top 10 products for July')
plt.show()
plt.clf()

top_products_of_april = top_products['Product_ID'].unique()
april_product_analysis = sales_2023[sales_2023['Product_ID'].isin(top_products_of_april)]
april_product_analysis_1 = april_product_analysis.groupby('Month')['Sales_Amount'].sum().reset_index().sort_values('Month')
sns.lineplot(x = april_product_analysis_1['Month'], y = april_product_analysis_1['Sales_Amount'])
plt.title('Revenue Trend of April Top 10 Products for 2023')
plt.show()
plt.clf()

# We can clearly see the top 10 performing products of April compared to their performance in other months as a group. April was a month where transactions were low, but revenue was high, which points towards the fact that products with higher profit margins performed well. Isolating the top 10 performing products in April and comparing their performance as a group across all months in 2023, we can see a large spike in the graph in April. This suggests the performance of these products generated high revenue with low transaction amounts.

april_product_analysis_2 = april_product_analysis.groupby(['State','Month'])['Sales_Amount'].sum().reset_index().sort_values('Month')

for state in april_product_analysis_2['State'].unique():
    subset = april_product_analysis_2[april_product_analysis_2['State'] == state]
    sns.lineplot(x = subset['Month'], y = subset['Sales_Amount'])
    plt.title(f"April top 10 product sales trend - {state}")
    plt.show()
    plt.clf()

# The top 10 performing products in April that suggested the increase in revenue for April have a consistent performance across the states. With no further data, I cannot investigate this further. However, my next steps would be understanding what was done differently regarding these months in April, whether it was a promotional activity, a brand deal or a marketing strategy that worked really well. This will allow the business to understand the underlying reasons behind these products' success, which as a result increased total revenue.

product_performance_july = sales_2023[sales_2023['Month'] == 7]
product_performance_july = product_performance_july.groupby('Product_ID')['Sales_Amount'].sum().sort_values(ascending = False).reset_index()
top_products_july = product_performance_july.head(10)
sns.barplot(x = top_products_july['Product_ID'], y = top_products_july['Sales_Amount'])
plt.title('Top 10 products for July')
plt.show()
plt.clf()

# Identifying the top 10 performing products of July.


top_products_of_july = top_products_july['Product_ID'].unique()
july_product_analysis = sales_2023[sales_2023['Product_ID'].isin(top_products_of_july)]
july_product_analysis_1 = july_product_analysis.groupby('Month')['Sales_Amount'].sum().reset_index().sort_values('Month')
sns.lineplot(x = july_product_analysis_1['Month'], y = july_product_analysis_1['Sales_Amount'])
plt.title('Revenue Trend of July Top 10 Products for 2023')
plt.show()
plt.clf()

# We can see this group of products had a large spike in July, this month is where the average sale per transaction was highest throughout the year.

july_product_analysis_2 = july_product_analysis.groupby(['State','Month'])['Sales_Amount'].sum().reset_index().sort_values('Month')

for state in july_product_analysis_2['State'].unique():
    subset = july_product_analysis_2[july_product_analysis_2['State'] == state]
    sns.lineplot(x = subset['Month'], y = subset['Sales_Amount'])
    plt.title(f"July top 10 product sales trend - {state}")
    plt.show()
    plt.clf()


# This group of products had a noticeable spike in each state for July, with all the revenue added up this is where we saw the highest sale average per transaction. NSW has 3 months where this group of products performed well and WA had this group of products peak in February, this was not noticeable on other graphs, as these products did not perform well in other states in February. Due to lack of data I cannot investigate this further, as stated for the month of April similarly I would analyse techniques used such as promotional activity, brand deals or marketing strategies within the months we see peaks in for these products.
```

## Are there seasonal patterns in product category performance?


```python
category_sales_monthly = sales_2023.groupby(['Month','Product_Category'])['Sales_Amount'].sum().reset_index().sort_values('Month')

for category in category_sales_monthly['Product_Category'].unique():
    subset = category_sales_monthly[category_sales_monthly['Product_Category'] == category]
    plt.plot(subset['Month'], subset['Sales_Amount'])
    plt.title(f'Monthly Sales of {category}')
    plt.xticks(range(1,13))
    plt.show()
    plt.clf()

# Revenue generated by the 4 product categories is highly variable, we see each category performing in separate 2-3 month periods. This separation suggests the influence of seasonality on revenue from different categories. This could also be a result of different promotional programs for each category or the influence of environmental factors on consumer behaviour.
```

##  How consistent is sales performance across states?

```python
state_sales_monthly = sales_2023.groupby(['Month','State'])['Sales_Amount'].sum().reset_index().sort_values('Month')

for state in state_sales_monthly['State'].unique():
    subset = state_sales_monthly[state_sales_monthly['State'] == state]
    plt.plot(subset['Month'], subset['Sales_Amount'])
    plt.title(f'Monthly Sales of {state}')
    plt.xticks(range(1,13))
    plt.show()
    plt.clf()

# Monthly sales across our operational states do not show a steady growth or decline over time. Considering our previous observations and insights, our data suggest there might be certain factors impacting the variability of our data. 

```
